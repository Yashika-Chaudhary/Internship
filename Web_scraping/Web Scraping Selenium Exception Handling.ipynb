# Question 2

## In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
producturls = []
brand_name= []
product_name =[]
price = []
retexchange = []
exp_delivery = []
avail = []


start = 0
end = 3
for page in range(start,end):
    url = driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url[0:10]:
        producturls.append(i.get_attribute("href"))
        
    next_button= driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    next_button.click()
    time.sleep(3)

for url in producturls[:20]:
    driver.get(url)
    time.sleep(2)
    
    try:
        brand_tags = driver.find_element(By.XPATH,'//*[@id="productOverview_feature_div"]/div/table/tbody/tr[1]/td[2]/span')
        brand_name.append(brand_tags.text)
    except NoSuchElementException:
        brand_name.append('-')

    try:
        product_tags = driver.find_element(By.XPATH,'//*[@id="productTitle"]')
        product_name.append(product_tags.text)
    except NoSuchElementException:
        product_name.append('-')


    try:
        price_tags = driver.find_element(By.XPATH,'//*[@id="corePriceDisplay_desktop_feature_div"]/div[1]/span[2]/span[2]/span[2]')
        price.append(price_tags.text)
    except NoSuchElementException:
        price.append('-')

    try:
        exchange_tags = driver.find_element(By.XPATH,'//*[@id="RETURNS_POLICY"]/span/div[2]/span')
        retexchange.append(exchange_tags.text)
    except NoSuchElementException:
        retexchange.append('-')

    try:
        delivery_tags = driver.find_element(By.XPATH,'//*[@id="mir-layout-DELIVERY_BLOCK-slot-PRIMARY_DELIVERY_MESSAGE_LARGE"]/span/span')
        exp_delivery.append(delivery_tags.text)
    except NoSuchElementException:
        exp_delivery.append('-')
        
    try:
        avail_tags = driver.find_element(By.XPATH,'//*[@id="availability"]/span')
        avail.append(avail_tags.text)
    except NoSuchElementException:
        avail.append('-')
        


print (len(brand_name), len(product_name), len(price), len(retexchange), len(exp_delivery), len(avail))

import pandas as pd
df = pd.DataFrame({'BRAND':brand_name, 'Product Name': product_name,'Price': price, 'Return/Exchange': retexchange, 'Expected Delivery': exp_delivery, 'Availibility': avail})
df
```

```python
guitar=df.to_csv('guitar.csv')
guitar
```

# Question 3

## Write a python program to access the search bar and search button on images.google.com and scrape 10images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’

```python
driver.get("https://images.google.com/")

keywords = ['guitar','cars','machine learning','cakes','fruits']

for key in keywords:
    img_urls =[]
    search_img = driver.find_element(By.CLASS_NAME,"gLFyf")
    search_img.send_keys(key)

    search_button = driver.find_element(By.CLASS_NAME,"Tg7LZd")
    search_button.click()
    time.sleep(3)
    
    for _ in range(2):
        driver.execute_script("window.scrollBy(0,500)")
        
    images = driver.find_elements(By.XPATH,'//img[@class="rg_i Q4LuWd"]')
    for image in images:
        source = image.get_attribute('src')
        if source is not None:
            if(source[0:4] == 'http'):
                img_urls.append(source)
    for i in range(len(img_urls)):
        if i >10:
            break
        print("Downloading {0} of {1} images of {2} ".format(i,10,key))
        response = requests.get(img_urls[i])
        file = open(r"/home/yashika/images"+str(key)+str(i)+".jpg","wb")
        file.write(response.content)
    driver.get("https://images.google.com/")
    time.sleep(3)
```

```python
comment = []
comment_time = []
comment_upvote = []

for _ in range(20):
    driver.execute_script("window.scrollBy(0,1000)")
    
    try:
        comment_tags = driver.find_elements(By.CLASS_NAME,'style-scope ytd-comment-renderer')
        for i in comment_tags[0:500]:
            comment.append(i.text.split('\n')[2])
            comment_time.append(i.text.split('\n')[1])
            comment_upvote.append(i.text.split('\n')[-2])
    except NoSuchElementException:
        comment.append('-')
        comment_time.append('-')
        comment_upvote.append('-')


print(len(comment ), len(comment_time), len(comment_upvote))

import pandas as pd
df = pd.DataFrame({'comment': comment, 'comment_time': comment_time, 'comment_upvote': comment_upvote})
df
```

## Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in“London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description. 

```python
driver.get("http://www.hostelworld.com/")

op = driver.find_element(By.XPATH,'//*[@id="search-field"]/div')
op.click()

loc = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div/div/div[4]/div/div[2]/div/div[1]/div/div/div/input')
loc.send_keys('London')

list = driver.find_element(By.XPATH,'//*[@id="predicted-search-results"]/li[2]/div')
list.click()

go = driver.find_element(By.XPATH,'//*[@id="search-button"]')
go.click()
```

```python
name = []
distance = []
rating =[]
review =[]
private_price = []
prices = []
link = []
des = []

url = driver.find_elements(By.XPATH,'//h2[@class= "title title-6"]/a')
for i in url:
    link.append(i.get_attribute("href"))

try:
    name_tags = driver.find_elements(By.XPATH,'//h2[@class= "title title-6"]')
    for i in name_tags:
        name.append(i.text)
except NoSuchElementException:
    name.append('-')

try:
    distance_tags = driver.find_elements(By.CLASS_NAME,"description")
    for i in distance_tags:
        distance.append(i.text)
except NoSuchElementException:
    distance.append('-')


try:
    rating_tags = driver.find_elements(By.CLASS_NAME,"score")
    for i in rating_tags[1:]:
        rating.append(i.text)
except NoSuchElementException:
    rating.append('-')

review =[]
try:
    review_tags = driver.find_elements(By.CLASS_NAME,"reviews")
    for i in review_tags:
        review.append(i.text)
except NoSuchElementException:
    review.append('-')


try:
    price_tags = driver.find_elements(By.CLASS_NAME,"price-col")
    for i in price_tags:
        prices.append(i.text)
except NoSuchElementException:
    prices.append('-')

private = []
dorm = []
private = prices[::2]
dorm = prices[1::2]


try:
    des_tags = driver.find_elements(By.CLASS_NAME,"info")
    for i in des_tags[1:]:
        des.append(i.text.split('\n')[:3])
except NoSuchElementException:
    des.append('-')

fac = []
try:
    fac_tags = driver.find_elements(By.CLASS_NAME,"facilities-label.facilities")
    for i in fac_tags:
        fac.append(i.text)
except NoSuchElementException:
    fac.append('-')

print(len(name), len(distance),len(rating),len(review),len(private),len(dorm),len(des),len(fac),len(link))

import pandas as pd
df = pd.DataFrame({'Name': name, 'distance': distance,'rating': rating, 'review': review,'Description':des, 'Facilities':fac, 'Private':private,'Dorm':dorm, 'Link':link})
df
```
